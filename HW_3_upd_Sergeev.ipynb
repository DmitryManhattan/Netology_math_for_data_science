### Задание 1

Дан объект в $2D$ пространстве

````python
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

При помощи линейных отображений:
- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x (поможет матрица `[[0, 1, 0], [1, 0, 0], [0, 0, 1]]`)


````python
# Уменьшить объект в два раза


B = np.array([[0.5,0,0],
              [0,0.5,0],
              [0,0,1]])

result = []

for i in A:
    result.append(B@i)


# Изобразим первоначальный объект
A = np.array(A)
x1 = A[:,0]
y1 = A[:,1]
plt.plot(x1, y1)

# Изобращим объект, который получили в результате преобразования
r = np.array(result)
x2 = r[:,0]
y2 = r[:,1]
plt.plot(x2, y2)

plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

````python
# Повернуть на 130 градусов

import math

# Здесь зададим на какой угол хотим повернуть
theta = np.radians(130)

C = np.array([[math.cos(theta),-math.sin(theta),0],
              [math.sin(theta),math.cos(theta),0],
              [0,0,0]])

result = []

for i in A:
    result.append(C@i)


# Изобразим первоначальный объект
A = np.array(A)
x1 = A[:,0]
y1 = A[:,1]
plt.plot(x1, y1)

# Изобращим объект, который получили в результате преобразования
r = np.array(result)
x2 = r[:,0]
y2 = r[:,1]
plt.plot(x2, y2)

plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

````python
# Отразить объект относительно прямой y=x (поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]])


D = np.array([[0,1,0],
              [1,0,0],
              [0,0,1]])

result = []

for i in A:
    result.append(D@i)


# Изобращим объект, который получили в результате преобразования
r = np.array(result)
x2 = r[:,0]
y2 = r[:,1]
plt.plot(x2, y2)

plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

Найдите спектральное разложение матрицы:

````python
m = np.array([[1, 2],
              [2, 3]])
print(m)
````

````python
v, w = np.linalg.eigh(m)
v
````

````python
# В функцию np.diag подставляем собственные значения матрицы m, чтобы получить диагональную матрицу

d = np.diag(v)
d
````

````python
# Получим матрицу S из собственных векторов

s = w.T
s
````

````python
# Получим матрицу St транспонировав матрицу S

st = s.T
st
````

````python
# Чтобы проверить перемножим 
st@d@s
````
