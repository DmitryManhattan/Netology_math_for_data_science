### Задание 1


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.


```python
import numpy as np
import math
import sympy as sym

# Создаем аналитический параметр
x = sym.Symbol('x')

some_function = x**5 + 4*sym.sin(2*x) + sym.cos(3*x+3)

# Находим первую производную
diff_1 = sym.diff(some_function, x)

print('Первая производная:')
diff_1
```

```python
# Значение первой производной в точке х = 1
diff_1.subs(x,1)
```

```python
diff_2 = sym.diff(diff_1, x)

print('Вторая производная:')
diff_2
```

```python
# Значение второй производной в точке х = 1
diff_1.subs(x,1)
```
